### Spectrogram

```python
import numpy as np
import librosa

def get_spectro(audio):
    audio, sr = librosa.load(audio)
    D = librosa.stft(audio) # see param
    S_db = librosa.amplitude_to_db(np.abs(D), ref=np.max)
    
    return S_db
```

```python
# Display Spectrogram
import os
import librosa.display
print(os.getcwd())
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(10,5))
S_db = get_spectro('../data/aldfly/XC479504.mp3')
img = librosa.display.specshow(S_db,
                               x_axis='time',
                               y_axis='linear',
                               ax=ax)

ax.set_title('Spectrogram', fontsize=20)
fig.colorbar(img, ax=ax, format=f'%0.2f')
plt.show()
```

```python
# Plot 5 premiers spectrograms
for i in range(1,6):
    plt.subplot(5, 1, i)
    librosa.display.specshow(get_spectro(audio_files[i-1]), 
                               x_axis='time', 
                               y_axis='log')
    plt.show()
```
